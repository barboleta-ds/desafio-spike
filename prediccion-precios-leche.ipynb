```python
# funcion para visualizar precipitaciones para un rango de fechas determinado
def time_plot_precipitation(region, fecha_inicial, fecha_final):
    columns = df_precipitaciones.columns
    if region not in columns:
        print("El nombre de la región no coincide. Por favor, intentalo con las siguientes etiquetas: {}".format(list(columns)))
    else:
        df2 = df_precipitaciones
        mask = (df2['date'] > fecha_inicial) & (df2['date'] <= fecha_final)
        df2 = df2.loc[mask]
        df2 = df2.set_index('date')
        #print(df2)

        plt.figure(figsize=(12,5))
        plt.title('Precipitaciones')
        plt.xlabel('Fecha')

        ax1 = df2[region].plot(color='blue', grid=True, label='mm caidos en {}'.format(region))

        h1, l1 = ax1.get_legend_handles_labels()
        plt.legend(h1, l1, loc=2)
```

Se puede apreciar, a grandes rasgos, que hay varias variables que tienen una alta correlacion entre ellas, lo que podría llevar a problemas por multicolinearidad. La multicolinearidad ocurre cuando hay dos o más variables independientes que tienen una alta correlación entre sí. Cuando algunas características están altamente correlacionadas, es posible que se presenten dificultades para distinguir entre sus efectos individuales sobre la variable dependiente. Además, el dataset tiene una alta dimensionalidad, por lo que es necesario disminuir la cantidad de variables que entrarán al modelo. Para ello, se utilizará el metodo de PCA, que me permitirá escoger una cantidad de variables significativas para aplicar en el modelo.

### PCA

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# lo primero es aislar la variable dependiente o target, la que que queremos predecir
y = df_modelo['Precio_leche']

# y luego crear el dataset solo de las variables (para analizar con PCA)
dataset_pca = df_modelo.drop(columns =["fecha", "Precio_leche"])

# normalizar y escalar el dataset para PCA
scaler = StandardScaler()
dataset_pca_normalized = pd.DataFrame(scaler.fit_transform(dataset_pca))
```

```python
# PCA
pca = PCA(n_components = 10) # se eligieron 10 componentes, porque un analisis previo mostró que 10 componentes eran los que explicaban gran parte de la variacion
pca.fit(dataset_pca_normalized,y)
pca_result = pca.transform(dataset_pca_normalized)

variance = pca.explained_variance_ratio_
var=np.cumsum(np.round(variance, decimals=3)*100)
print("Porcentaje acumulado de variación de cada componente")
var 
```

```python
# visualizacion de los dos primeros componentes
def pca_plot(score,coeff,labels=None):
    xs = score[:,0]
    ys = score[:,1]
    n = coeff.shape[0]

    plt.scatter(xs ,ys, c = y) #without scaling
    for i in range(n):
        plt.arrow(0, 0, coeff[i,0], coeff[i,1],color = 'r',alpha = 0.5)
        if labels is None:
            plt.text(coeff[i,0]* 1.15, coeff[i,1] * 1.15, "Var"+str(i+1), color = 'g', ha = 'center', va = 'center')
        else:
            plt.text(coeff[i,0]* 1.15, coeff[i,1] * 1.15, labels[i], color = 'g', ha = 'center', va = 'center')
    
    plt.xlabel("PC{}".format(1))
    plt.ylabel("PC{}".format(2))
    plt.grid()

pca_plot(pca_result[:,0:2], pca.components_)
```

```python
# scores de cada variable en cada componente
df_pca_scores = pd.DataFrame(pca.components_, columns=list(dataset_correlacion.columns))
df_pca_scores
```

```python
# variables con mayor peso en cada componente
n_pcs= pca.components_.shape[0]
most_important = [np.abs(pca.components_[i]).argmax() for i in range(n_pcs)]
initial_feature_names = list(dataset_correlacion.columns)
most_important_names = [initial_feature_names[most_important[i]] for i in range(n_pcs)]
dic = {'PC{}'.format(i): most_important_names[i] for i in range(n_pcs)}
dic
```

En ese ultimo dic, se aprecian las variables que tienen mayor peso en cada dimension. Se tomarán en cuenta 9 de ellas, exluyendo la "No_sabe__No_responde_Miles_de_personas" porque intuyo (solo por el nombre) que es una variable que no aporta ninguna información.
